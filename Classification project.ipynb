```python
data = pd.read_csv('D:\\College\\Senior-2 CSE 2\\Big Data\\uci-open-university-learning-analytics-dataset\\merged2001.csv')
```

```python
y_soft = data['new score']
column_indexes = list(range(14, 208))
X_soft = data.iloc[:,column_indexes].astype(int)
```

```python
import tensorflow as tf
y_train_onehot = tf.keras.utils.to_categorical(y_soft, num_classes=10)
```

```python
# Splitting the dataset into the Training set and Test set  
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X_soft, y_train_onehot, test_size = 0.25, random_state = 0)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
softmax_model = Sequential()
softmax_model.add(Dense(units=512, activation='relu', input_dim=len(X_train.columns)))
softmax_model.add(Dense(units=1024, activation='relu'))
softmax_model.add(Dense(units=512, activation='relu'))
softmax_model.add(Dense(units=10, activation='softmax'))
```

```python
softmax_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python

softmax_model.fit(X_train, y_train_onehot, epochs=10, batch_size=32, validation_split=0.2)
```

```python
loss, accuracy = softmax_model.evaluate(X_train, y_train)
print("Accuracy:", accuracy)
```
